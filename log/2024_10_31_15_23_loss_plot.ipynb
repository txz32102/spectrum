```python
import numpy as np
import matplotlib.pyplot as plt

def read_loss(file_path):
    epochs = []
    losses = []
    
    with open(file_path, 'r') as file:
        for line in file:
            parts = line.split(", ")
            if len(parts) == 2:
                epoch = int(parts[0].split('[')[1].split('/')[0])
                loss = float(parts[1].split(": ")[1])
                epochs.append(epoch)
                losses.append(loss)
    
    return np.array(epochs), np.array(losses)

# Paths to your log files
log_file_1 = '2024_10_31_loss_resnet34+attention.txt'
log_file_2 = '2024_10_31_loss_resnet34.txt'  # Replace with the actual path to your second log file

# Read losses
epochs_1, losses_1 = read_loss(log_file_1)
epochs_2, losses_2 = read_loss(log_file_2)

# Plotting
plt.figure(figsize=(12, 6))
plt.plot(epochs_1, losses_1, label='ResNet34 + Attention', color='blue')
plt.plot(epochs_2, losses_2, label='Second Model', color='orange')
plt.title('Training Loss Comparison')
plt.xlabel('Epoch')
plt.ylabel('Average Loss')
plt.yscale('log')  # Use logarithmic scale if there are large differences in loss values
plt.legend()
plt.grid()
plt.savefig("1.png")
```
